```python
# Data Expolarity and imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the Producer prices dataset
# df = pd.read_csv('FAOSTAT_IRAN_POPULATION.csv')
# Use raw string to avoid escaping backslashes
file_path = r"FAOSTAT_IRAN_PRODUCER_PRICES.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Show first few rows to verify
print(df.head())

# Display the first few rows of the dataset
print(df.head())
print(df.shape)
```

```python
# Initial Data Exploration

# Display the first 5 rows
print("First 5 rows:")
print(df.head())

# Get dataset information (data types, non-null counts)
print("\nDataset Info:")
print(df.info())

# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values in each column
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
#EDA

# Univariate Analysis

# Distribution of Products (Items)

plt.figure(figsize=(10, 12))
sns.countplot(y='Item', data=df, order=df['Item'].value_counts().index)
plt.title("Distribution of Products (Items)")
plt.xlabel("Count")
plt.ylabel("Item")
plt.show()


# Distribution of Years

plt.figure(figsize=(10, 6))
sns.countplot(x='Year', data=df, order=df['Year'])
plt.title("Distribution of Years in Producer Prices Data")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#EDA

# Bivariate Analysis

# Scatter plot: Price (Value) vs. Year

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Year', y='Value', data=df)
plt.title("Producer Price vs. Year")
plt.xlabel("Year")
plt.ylabel("Price (Value)")
plt.xticks(rotation=45)
plt.show()


# Boxplot: Price (Value) distribution by Item

plt.figure(figsize=(12, 6))
sns.boxplot(x='Item', y='Value', data=df)
plt.title("Price (Value) Distribution by Item")
plt.xlabel("Item")
plt.ylabel("Price (Value)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Data Cleaning for Producer Prices Dataset

# Check for missing values before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# Handling Missing Values
# Fill missing 'Value' (price) with the median Value
df['Value'] = df['Value'].fillna(df['Value'].median())

# Drop rows where 'Item' or 'Area' is missing (Area is always Iran in this dataset, but good practice)
df = df.dropna(subset=['Item', 'Area'])

# Drop duplicate rows
print("\nNumber of duplicate rows before cleaning:", df.duplicated().sum())
df = df.drop_duplicates()
print("Number of duplicate rows after cleaning:", df.duplicated().sum())

# Verify missing values after cleaning
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Data Preprocessing for Producer Prices Dataset

# Encode 'Item' using LabelEncoder
le = LabelEncoder()
df['Item_encoded'] = le.fit_transform(df['Item'])

# One-hot encode the 'Area' column (kept for structural consistency)
df = pd.get_dummies(df, columns=['Area'], drop_first=True)

# Drop original 'Item' column
df = df.drop(columns=['Item'])

# Display a sample of the dataset after preprocessing
print("Data after preprocessing:")
print(df.head())
```

```python
# Data Normalization/Standardization for Producer Prices Dataset

# Initialize StandardScaler
scaler = StandardScaler()

# Standardize the 'Value' column
# Note: We are only scaling 'Value' as 'Year' is a temporal feature and 'Item_encoded' is already categorical.
df[['Value']] = scaler.fit_transform(df[['Value']])

# Verify the changes with a summary statistic for the scaled column
print("Summary after scaling 'Value':")
print(df[['Value']].describe())
```

```python
# Final overview and sanity check

print("Final dataset shape:", df.shape)
print("Columns in final dataset:", df.columns)
print("\nA sample of the final processed data:")
print(df.head())
```
