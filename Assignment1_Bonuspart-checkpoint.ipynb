```python
import requests
import pandas as pd

# Define keys (column names)
specs_keys = ['volume', 'engine', 'acceleration', 'fuel']
price_keys = ['price']
detail_keys = ['url', 'title', 'year', 'mileage', 'transmission', 'color', 'description']

# Create empty DataFrame
df = pd.DataFrame(columns=detail_keys + specs_keys + price_keys)


def add_to_df(cars, df):
    for car in cars:
        try:
            detail = car.get('detail', {})

            # Skip old cars
            # Assuming 'year' in detail might be in Shamsi (Persian calendar)
            # If it's Gregorian, this condition needs adjustment
            year_str = detail.get('year', '0')
            try:
                # Attempt to convert Shamsi year to integer
                year_int = int(year_str)
            except ValueError:
                # Handle cases where year might not be a simple integer (e.g., empty string, bad format)
                year_int = 0 # Assign a default that will likely be skipped

            if year_int < 1390:
                continue

            record = []

            # Add detail info
            for key in detail_keys:
                record.append(detail.get(key, None))

            # Add specs info
            specs = car.get('specs', {})
            for key in specs_keys:
                record.append(specs.get(key, None))

            # Add price info
            price = car.get('price', {})
            for key in price_keys:
                record.append(price.get(key, None))

            df.loc[len(df)] = record

        except Exception as e:
            print("Skipping a record due to error: {}".format(e))

    return df


# Scrape pages
for i in range(1, 4):
    try:
        response = requests.get(
            'https://bama.ir/cad/api/search?vehicle=samand%2Clx&pageIndex={}'.format(i),
            headers={"User-Agent": "Mozilla/5.0"}
        )
        response.raise_for_status()

        data = response.json()
        cars = data['data']['ads']
        df = add_to_df(cars, df)

    except Exception as e:
        print("Error on page {}: {}".format(i, e))


# Save dataset
df.to_csv('samand_data.csv', index=False, encoding='utf-8')

print("Data saved successfully as samand_data.csv")
print(df.head())
```
