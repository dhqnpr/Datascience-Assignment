```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_selection import mutual_info_classif
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Style
sns.set(style="whitegrid", context="talk")

# Load Producer Prices Dataset
file_path = r"FAOSTAT_IRAN_PRODUCER_PRICES.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Display first few rows and basic info
print("Dataset shape:", df.shape)
print(df.head())
df.info()
```

```python
# Check missing values on the entire dataset
print("Missing values per column:\n", df.isnull().sum())

#clean data (drop missing Value)
df = df.dropna(subset=['Value'])

#take a sample
df_sample = df.sample(n=10000, random_state=42) if len(df) > 10000 else df

#Plot distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df_sample['Value'], kde=True, ax=axes[0])
axes[0].set_title("Producer Price Value Distribution")

sns.histplot(df_sample['Year'], kde=True, ax=axes[1])
axes[1].set_title("Year Distribution")

plt.tight_layout()
plt.show()

#Boxplot of Value by Item 
#It is vertical because of the large amount of data
plt.figure(figsize=(10, 25))
sns.boxplot(y='Item', x='Value', data=df_sample)

plt.title("Producer Price Value Distribution by Item")
plt.tight_layout()
plt.show()

```

```python
# Create a synthetic feature for demonstration

# Create a "Price_Index" by scaling producer price values per item
df['Price_Index'] = df.groupby('Item')['Value'].transform(lambda x: x / x.mean())

# Add a synthetic "Estimated_Cost" — assume e.g. 70% of the recorded price value
df['Estimated_Cost'] = df['Value'] * 0.7

# Calculate "Estimated_Profit" = Value - Cost
df['Estimated_Profit'] = df['Value'] - df['Estimated_Cost']

# Profit–to–Cost ratio
df['Profit_Cost_Ratio'] = df['Estimated_Profit'] / df['Estimated_Cost']

# Bin prices into quantile categories — Low / Medium / High / Very High
df['Price_Bin'] = pd.qcut(df['Value'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Square root transformation on Value (sometimes used to reduce skewness)
df['Sqrt_Value'] = np.sqrt(df['Value'])

# Combine columns — for easy referencing
# Create a composite ID joining Year and Item name
df['Year_Item'] = df['Year'].astype(str) + "_" + df['Item'].astype(str)

# Show sample of new features
df[['Item', 'Year', 'Value', 'Estimated_Cost', 'Estimated_Profit', 
    'Profit_Cost_Ratio', 'Price_Bin', 'Sqrt_Value', 'Year_Item']].head()
```

```python
# Convert 'Year' to a datetime object (representing Jan 1 of that year)
df['Date'] = pd.to_datetime(df['Year'], format='%Y')

# Extract Year directly (redundant but keeps same naming pattern)
df['Year'] = df['Date'].dt.year

# Extract month and day (synthetic, since FAOSTAT data is annual)
# We'll fill each record as if it represents data for January
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Create 'DayOfWeek' (the weekday of Jan 1 for each year — just descriptive)
df['DayOfWeek'] = df['Date'].dt.day_name()

# Display sample of date-related features
df[['Date', 'Day', 'Month', 'Year', 'DayOfWeek']].head()
```

```python
# Aggregation by Item (product)
agg_item = df.groupby('Item').agg({
    'Year': 'nunique',      # how many years this item appears
    'Value': ['mean', 'sum', 'max', 'min']   # price metrics
}).reset_index()

# Rename columns for readability
agg_item.columns = ['Item', 'Year_Count', 'Avg_Price', 'Total_Price', 'Max_Price', 'Min_Price']

print("Aggregated producer price metrics by item:")
print(agg_item.head())

# Create yearly average price dataframe
yearly_prices = (
    df.groupby('Year')['Value']
      .mean()
      .reset_index(name='Avg_Price')
)

# Visualization: Average Producer Price across Years
plt.figure(figsize=(10,5))
sns.lineplot(data=yearly_prices, x='Year', y='Avg_Price', marker='o', color='tab:blue')
plt.title("Average Producer Price Over Years")
plt.xlabel("Year")
plt.ylabel("Average Producer Price (Value)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Create binary target: HighPrice
median_price = df['Value'].median()
df['HighPrice'] = (df['Value'] > median_price).astype(int)

# Feature engineering
df['Price_Mean_By_Item'] = df.groupby('Item')['Value'].transform('mean')
df['Price_Range_Item'] = df.groupby('Item')['Value'].transform(lambda x: x.max() - x.min())

#Select numerical features
features = [
    'Year',
    'Value',
    'Price_Mean_By_Item',
    'Price_Range_Item'
]

X = df[features].fillna(0)
y = df['HighPrice']

# Compute Mutual Information
mi_scores = mutual_info_classif(X, y, random_state=42)
mi_series = pd.Series(mi_scores, index=features).sort_values(ascending=False)

print("Mutual Information scores for numerical features:")
print(mi_series)
```

```python
# Standardize the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (2 components for visualization)
pca = PCA(n_components=2, random_state=42)
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with PCA results and target variable (HighPrice)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['HighPrice'] = y.values

# Explained variance
print("Explained Variance Ratio by PCA components:", pca.explained_variance_ratio_)

# Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df,
    x='PC1', y='PC2',
    hue='HighPrice',
    palette='coolwarm',
    alpha=0.7
)
plt.title("PCA of FAOSTAT Producer Price Features")
plt.show()
```

## When is Feature Engineering a “Nice to Have” vs. a “Must to Have”?

Feature engineering is a **"nice to have"** when:
- The raw data is already clean, well-structured, and highly informative.
- Modern models (such as deep learning architectures) can automatically learn complex, non-linear patterns from large datasets.
- Baseline model already performs well and any further gains are marginal.

On the other hand, feature engineering becomes a **"must to have"** when:
- The raw data is noisy, unstructured, sparse, or lacks explicit domain signals.
- Model performance is clearly constrained by the quality and expressiveness of the input features.
- Inject domain expertise to design new features that meaningfully improve predictive power.
- Have limited data or compute, and need compact, efficient, and interpretable features to train models effectively.

To sum up, for messy, complex, or heterogeneous real-world data—such as many retail transaction datasets—feature engineering often plays a critical role in uncovering hidden structure and achieving substantial performance improvements.
